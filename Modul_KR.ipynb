```python
#check for duplicates in df

duplicates = df.duplicated()

df['IsDuplicate'] = duplicates

num_duplicates = df.duplicated().sum()

print("Number of Duplicate Rows:", num_duplicates)

print(df)
```

```python
#check for na

na_counts = df.isna().sum()

print("Number of NaN values in each column:")
print(na_counts)
```

Якщо є пропуски, замінить тип даних на float і середнє значення за
стовпцем.

```python
#change the na values
for col in df.columns:

    if pd.api.types.is_numeric_dtype(df[col]) and df[col].isna().any():
        mean_value = df[col].mean()
        df[col] = df[col].fillna(mean_value)

print("DataFrame after replacing NaN values with the mean:")
print(df)
```

Ще раз перевірте кількість пропусків

```python
#check for na again
na_counts = df.isna().sum()

print("Number of NaN values in each column:")
print(na_counts)
```

Знайдіть країну, де найбільше відео було завантажено на YouTube

```python
uploads_by_country = df.groupby('Country')['uploads'].sum()

# Find the country with the most video uploads
country_with_most_uploads = uploads_by_country.idxmax()
most_uploads_count = uploads_by_country.max()
print("the country with most uploads is",country_with_most_uploads, "with",  most_uploads_count, "uploads")
```

Знайдіть відео (виведіть його назву) з найбільшою та найменшою
кількістю завантажень uploads

```python
# find the youtuber with most viewes
views_by_youtuber = df.groupby('Youtuber')['video views'].sum()

youtuber_with_most_views = views_by_youtuber.idxmax()
most_views_count = views_by_youtuber.max()

print(f"YouTuber with the most views: {youtuber_with_most_views}")
print(f"Number of views: {most_views_count}")
```

```python
import plotly.graph_objects as go
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris_data = iris.data
iris_target = iris.target


fig = go.Figure()

for species in range(3):
    species_data = iris_data[iris_target == species]
    fig.add_trace(go.Scatter(
        x=species_data[:, 1],
        y=species_data[:, 0],
        mode='markers',
        marker=dict(
            size=species_data[:, 2],
            color=species_data[:, 3],
            colorscale='Viridis',
            colorbar=dict(title='Petal Width')
        ),
        text=f"Petal Width: {species_data[:, 3]}",
        name=f'Species {species}'
    ))


fig.update_layout(
    title='Scatter Plot for Iris Dataset',
    xaxis=dict(title='Sepal Width'),
    yaxis=dict(title='Sepal Length'),
    hovermode='closest'
)

fig.show()
```
